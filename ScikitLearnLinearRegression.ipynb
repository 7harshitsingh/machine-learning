
# **SciKit Learn Closed Linear Regression**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('/content/CarPrice_Assignment.csv')
df.head()
```

```python
features = df[['horsepower', 'highwaympg', 'peakrpm', 'curbweight', 'wheelbase', 'stroke']]

X_train = np.array(features)
y_train = np.array(df['price'])

scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)

print(X_norm)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_norm, y_train)

b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
```

```python
X_predict = [[ 0.179, -0.547, -0.223, -0.014, -1.86, -1.34], [ 0.140, -0.546, -0.889, -0.1109, -1.27, -1.83], [ 1.261, 0.6891,  0.2623, 0.5137, -0.70,  0.6999]]

prediction = linear_model.predict(X_predict)
print(f'prediction using scikit-learn linear regression : {prediction}')
```
