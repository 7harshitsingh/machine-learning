## TensorFlow Keras

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

The CSV files contains ---
- other columns represents various features on which the result depends
- smoking column which indicates Y/N for smokers and non-smokers


```python
df = pd.read_csv("/content/smoking.csv")

map = {'Y':1,'N':0}
gender = {'M':1, 'F':0}

df['gender'] = df['gender'].map(gender)
df['oral'] = df['oral'].map(map)
df['tartar'] = df['tartar'].map(map)

df.head(5)
```

```python
X = np.array(df.loc[:, ~df.columns.isin(['ID', 'smoking'])])
y = np.array(df[["smoking"]])

m=X.shape[0]

print(X.shape)

X_train = X[:int(m*0.95)]
y_train = y[:int(m*0.95)]

print(X_train.shape)

X_test = X[:m-int(m*0.95)]
y_test = y[:m-int(m*0.95)]

print(X_test.shape)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(25,)),
    tf.keras.layers.Dense(units=10, activation="relu", name="layer1"),
    tf.keras.layers.Dense(units=4, activation="relu", name="layer2"),
    tf.keras.layers.Dense(units=1, activation="sigmoid", name="layer3"),
])

model.compile(loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print(f"Loss - {loss}")
print(f"Accuracy - {accuracy}")
```
